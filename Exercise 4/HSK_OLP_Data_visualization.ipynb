```python
import pandas as pd 
import numpy as np
import time
import datetime
```

```python
data_frame = pd.read_csv('data_file.csv', index_col=0, parse_dates=True, infer_datetime_format=True)
print(data_frame.head())
data_frame.info()
```

```python
data_frame.resample('D').sum()
data_frame=data_frame.resample('D').sum()
```

```python
data_frame.index
data_frame.head()
data_frame.reset_index(level=0, inplace=True)
data_frame.tail(10)
```

```python
data_frame['CF_216'] = data_frame['216'].cumsum()
data_frame['CF_5468'] = data_frame['5468'].cumsum()
data_frame['CF_2947'] = data_frame['2947'].cumsum()
data_frame['CF_2483'] = data_frame['2483'].cumsum()
data_frame['CF_1300'] = data_frame['1300'].cumsum()
data_frame['CF_3215'] = data_frame['3215'].cumsum()
data_frame['CF_6264'] = data_frame['6264'].cumsum()
data_frame['CF_4488'] = data_frame['4488'].cumsum()
data_frame['CF_7330'] = data_frame['7330'].cumsum() 
```

```python
data_frame.to_csv(r'cumulative_sum_data_OLPE_HSK_.csv', encoding='utf-8', index=False)
data_frame.tail(10)
```
